```
#load numpy and panda libraries
import numpy as nm
import pandas as pd
import matplotlib.pyplot as plt
import missingno as msno
import seaborn as sns

df=pd.read_csv('https://raw.githubusercontent.com/Michael-C-1/Module10/main/Module10%20data%20set.csv')
df.head()
```

```
#check for nulls
df.isnull().sum()
```

```
#lists unique number values
df.nunique()
```

```
df.describe()
```

```
for col in df:
  sns.kdeplot(df[col], shade=True)
```

```
normalized_df=(df-df.mean())/df.std()
for col in df:
  sns.kdeplot(normalized_df[col], shade=True)
```

```
min_max_df=(df-df.min())/(df.max()-df.min())
for col in df:
  sns.kdeplot(min_max_df[col], shade=True)
```

```
with sns.axes_style('white'):
  sns.jointplot("T", "EP", df, kind='kde')
  sns.jointplot("AP", "EP", df, kind='kde')
  sns.jointplot("RH", "EP", df, kind='kde')
  sns.jointplot("V", "EP", df, kind='kde')
```

```
df.hist(bins=50, figsize=(15,10))
plt.show()
```

```
corr=df.corr()
fig, ax=plt.subplots(figsize=(10,10))
sns.heatmap(data=corr, square=True, annot=True, cbar=True, ax=ax)
```
